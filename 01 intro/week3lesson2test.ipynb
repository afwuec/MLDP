```python
## Read data
import pandas as pd
FILE_PATH= "iris.csv"
df = pd.read_csv(FILE_PATH)
df ## Display dataframe
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['Species'], axis=1).to_numpy()
y = df['Species'].to_numpy()
df

test_size = 0.3
random_state = 3000
X_train, X_test, y_train, y_test = train_test_split(X,
y,
test_size=test_size,
random_state=random_state)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
## Initialise the models
logr = LogisticRegression(max_iter=200)
lda = LinearDiscriminantAnalysis()
logr.fit(X_train, y_train)
## Evaluate the Model
## Make predictions on the test set using X_test
y_pred = logr.predict(X_test)
## Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
print("Logistic Regression: ", accuracy_score(y_test, y_pred))


lda.fit(X_train, y_train)
## Evaluate the Model
## Make predictions on the test set using X_test
y_pred = lda.predict(X_test)
## Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
print("linear discriminant analysis: ", accuracy_score(y_test, y_pred))
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
FILE_PATH= "gpa.xlsx"
df = pd.read_excel(FILE_PATH)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                   
df ## Display dataframe
```

```python
col_y='Overall'
y=df[col_y]
col_irrelevant=[col_y,'Department', 'Semester']
X=df.drop(col_irrelevant, axis=1)
X=pd.get_dummies(X, drop_first=True)
X 
```

```python
from sklearn.model_selection import train_test_split
test_size = 0.3
random_state = 2025
X_train, X_test, y_train, y_test = train_test_split(X,
y, test_size=test_size, random_state=random_state)


```

```python
from sklearn.linear_model import LinearRegression
linr=LinearRegression(fit_intercept=True)
linr.fit(X_train, y_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
## Initialise the models
logr = LogisticRegression(max_iter=200)
lda = LinearDiscriminantAnalysis()
logr.fit(X_train, y_train)
## Evaluate the Model
## Make predictions on the test set using X_test
y_pred = logr.predict(X_test)
## Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
print("Logistic Regression: ", accuracy_score(y_test, y_pred))


lda.fit(X_train, y_train)
## Evaluate the Model
## Make predictions on the test set using X_test
y_pred = lda.predict(X_test)
## Calculate the accuracy of the model
from sklearn.metrics import accuracy_score
print("linear discriminant analysis: ", accuracy_score(y_test, y_pred))
```
